```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = np.genfromtxt("./datasets/DiscreteSpectrumExample/DiscreteSpectrumExample_train1_x.csv", delimiter=",", dtype=np.float32)
x = data.reshape(-1, 51, 2)
```

```python
for i in np.random.permutation(len(x))[:50]:
    plt.scatter(x[i, :, 0], x[i, :, 1], marker=',', s=1)
plt.axhline(y=0)
plt.axvline(x=0)
plt.show()
```
